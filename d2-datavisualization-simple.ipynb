**Visualizing sales trends with python using sample dataset**

**Step 1: Import Required Libraries**

Start by importing essential libraries.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries imported succesfully 3/3")
```

**Step 2: Load the Sample Dataset**

We'll use a sample dataset. You can create your own or use the example provided below. Save the dataset as sales_data.csv.

```python
# Load dataset
df = pd.read_csv('sales_data.csv')

# Display the first few rows
print(df.head())
```

**Step 3: Data Pre-processing**

Ensure the data is clean and suitable for analysis.


```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Aggregate sales by date
sales_trend = df.groupby('Date')['Sales'].sum().reset_index()
print(sales_trend)
```

**Step 4: Data Visualization**

Use Matplotlib and Seaborn to create meaningful visualizations.


```python
print("This is line chart")
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_trend, x='Date', y='Sales', marker='o', color='blue')
plt.title('Sales Trend Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(True)
plt.show()
```

```python
print("This is bar chart")
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product', y='Sales', ci=None, palette='viridis')
plt.title('Sales by Product', fontsize=16)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.show()
```

```python
print("This is heatmap")
# Example: If dataset has other numerical features
# Compute correlation
correlation = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
